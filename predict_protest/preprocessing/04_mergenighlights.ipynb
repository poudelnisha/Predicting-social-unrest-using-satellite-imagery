## Merge Protest data with nightlights

### Before running this notebook, please run, 

1. 02mapprotest_pixel.ipynb. You will get 2feballevents.csv 
2. Download a landsat image, (a tile that you want to experiment with. Here I do the selective sampling and 
   take the tile in an around of capital city of Nepal,where the number of protest is relatively high)  (Use this step if you are using rasters in addition to nightlights otherwise skip this step)
3. Download nightlights from google earth engine using 03Downloadviirsdnb_nightlights.ipynb (you can download nightlights from https://eogdata.mines.edu/products/vnl/#v1 
    Here we are using   VNL_v2_npp_2018_global_vcmslcfg_c202101211500.average.tif.gz. It's detail is given in the website).You can download the nightlights from your Area of interest 
    if you use Google Earth Engine(GEE).
4. Crop the nightlights tif as per the landsat image we have downloaded. (As I have plans for further using this   nightlights.tif with other raster,
   I adopt this approach.However, we can read corresponding nightlights with the longitude and lattitude.Pixelwise raster reading code is given in tiftocsv.txt inside nightlights folder. we can read the raster as per        desired resolution also.
   
**Input** : Cropped nightlight raster. nightlights/nightlights.tif  

**Output**: merged file with nightlights and protestcount. 1marfinal_datatrainig.csv"

```python
from osgeo import gdal
dem = gdal.Open('../cropped/electricityvalue.tif')
gt = dem.GetGeoTransform()
gt
```

```python
dem.RasterXSize,dem.RasterYSize
```

```python
import shapely
# get coordinates of upper left corner
xmin = gt[0]
ymax = gt[3]
res = gt[1]

# determine total length of raster
xlen = res * dem.RasterXSize
ylen = res * dem.RasterYSize

# number of tiles in x and y direction
xdiv = 156
ydiv = 176

# size of a single tile
xsize = xlen/xdiv
ysize = ylen/ydiv

# create lists of x and y coordinates
xsteps = [xmin + xsize * i for i in range(xdiv+1)]
ysteps = [ymax - ysize * i for i in range(ydiv+1)]
grid_cell = []
# loop over min and max x and y coordinates
centroid_grid =[]
crs = "EPSG:4326"
for i in range(xdiv):
    for j in range(ydiv):
        xmin = xsteps[i]
        xmax = xsteps[i+1]
        ymax = ysteps[j]
        ymin = ysteps[j+1]
        center_x = (xmax-xmin)/2
        center_y = (ymax-ymin)/2
        grid_cell.append( shapely.geometry.box(xmin, ymin, xmax, ymax) ) 
        centroid_grid.append((xmin+center_x,ymin+center_y))

tif_cell = geopandas.GeoDataFrame(grid_cell, columns=['geometry'], 
                                 crs=crs)
#center_cell = pd.DataFrame(centroid_grid,columns=['long_lat'])
#         print("xmin: "+str(xmin))
#         print("xmax: "+str(xmax))
#         print("ymin: "+str(ymin))
#         print("ymax: "+str(ymax))
#         print("\n")
        
# close the open dataset!!!
dem = None       
 

```
